## Exploring Dataset (Finding Statistical Info)

```python
# Explore the dataset
# print(df.info())
print(df.describe())

# Handle missing values (if any)
df = df.dropna()

```

## Feature Selection and Feature Scaling

```python
# Scale the numerical features
scaler = StandardScaler()
X = scaler.fit_transform(df.drop('Class', axis=1))
y = df['Class']

```

## Train Test Split

```python
# Split the dataset into training, and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)


```

## Model Selection and Training (Fitting the Train Set into the Model)

## LOGISTIC REGRESSION

```python
# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

```

## Making Predictions on the Test Set

```python
# Test Set
# Model Evaluation
# Logistic Regression
lr_test_acc = accuracy_score(y_test, lr_model.predict(X_test))
lr_test_precision = precision_score(y_test, lr_model.predict(X_test))
lr_test_recall = recall_score(y_test, lr_model.predict(X_test))
lr_test_f1 = f1_score(y_test, lr_model.predict(X_test))

print("Logistic Regression Test Metrics:")
print(f"Accuracy: {lr_test_acc:.4f}")
print(f"Precision: {lr_test_precision:.4f}")
print(f"Recall: {lr_test_recall:.4f}")
print(f"F1-score: {lr_test_f1:.4f}")

```

#### The Accuracy is very high, as the same as the Precision, but for the Recall and F1-Score the values are not that low but it will be boosted later

### Calibrating the Logistic Regression Model with Isotonic Regression for Optimal Performance

```python
from sklearn.calibration import CalibratedClassifierCV

# Logistic Regression with Isotonic Regression calibration
lr_model = LogisticRegression(class_weight='balanced')
lr_model.fit(X_train, y_train)

# Calibrate the Logistic Regression model using Isotonic Regression
calibrated_lr_model = CalibratedClassifierCV(lr_model, method='isotonic', cv='prefit')
calibrated_lr_model.fit(X_train, y_train)

y_pred_proba = calibrated_lr_model.predict_proba(X_test)
y_pred = (y_pred_proba[:, 1] > 0.5).astype(int)  # Adjust the threshold as needed

lr_test_acc = accuracy_score(y_test, y_pred)
lr_test_precision = precision_score(y_test, y_pred)
lr_test_recall = recall_score(y_test, y_pred)
lr_test_f1 = f1_score(y_test, y_pred)

print("Logistic Regression Test Metrics (Calibrated with Isotonic Regression):")
print(f"Accuracy: {lr_test_acc:.4f}")
print(f"Precision: {lr_test_precision:.4f}")
print(f"Recall: {lr_test_recall:.4f}")
print(f"F1-score: {lr_test_f1:.4f}")

print("\n")

print("Confusion Matrix for Test Set:")
con_mat_lr = confusion_matrix(y_test, lr_model.predict(X_test))
print(con_mat_lr)
```

#### Here's the Performance Metrics on the Test Set showing that there's Improvement compared to the previous Result Metrics.
